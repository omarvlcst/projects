# Ecuaciones de Estructura para Estrellas Maximalmente Compactas

De las ecuaciones relativistas de equilibrio hidrostático para estrellas (Tolman-Oppenheimer-Volkoff):

$$\frac{dp}{dr} = - \dfrac{G}{c^2} \dfrac{(p+\varepsilon)\left(m+\frac{4\pi r^3p}{c^2}\right)}{r\left(r-\frac{2Gm}{c^2}\right)}$$

$$\frac{dm}{dr} = 4\pi r^2 \dfrac{\varepsilon}{c^2}$$

De la hipótesis fuertemente solicitada por la causalidad, defínase la cantidad $s$ tal que: 

$$\dfrac{dp}{d\epsilon}=\dfrac{c_s^{\,\,2}}{c^2}=s^2\leq1,$$

donde $c_s$ es la velocidad de sonido del fluido en la estrella. Para el límite casual $s=1$, se propone así una ecuación de estado en la forma:

$$ p = 0 \text{ para } \varepsilon \leq \varepsilon_0$$

$$ p = \varepsilon-\varepsilon_0 \text{ para } \varepsilon > \varepsilon_0$$

donde $\epsilon_0$ es un parámetro libre que proviene de la constante de integración de la ecuación $dp\,/\,d \epsilon=s=1$, que equivale al valor de la densidad finita en el radio de la estrella $r=R^*$, para el cual la presión se anula. Tomándolo en cuenta, las ecuaciones de estructura pueden ser reescritas en una versión adimensional:

$$\frac{dw}{dx} = - \dfrac{(2w-1)\left(y+4\pi x^3(w-1)\right)}{x\left(x-2y\right)}$$

$$\frac{dy}{dx} = 4\pi x^2 w$$

para la nueva variable $\omega=\frac{\varepsilon}{\varepsilon_0}$

```python
# Constantes Universales
G = 6.6743015e-20 #km³/kg*s²
c = 299792.458 #km/s
Msol = 1.98847e30 #kg
Masas_solares = G*Msol/c**2 #en unidades geométricas (km)
a = 0.000001
b = 0.25
N = 3000
h = (b-a)/float(N)
w0=3.034 #Valor critico maximo
```

```python
def f(r,x):
    w = r[0]
    y = r[1]
    fw = -(2.*w-1)*(y+4.*np.pi*x**3.*(w-1.))/(x*(x-2.*y))
    fy = 4.*np.pi*x**2.*w
    return np.array([fw,fy],float)

def RK4(f,w_init):
    x = np.arange(a,b,h)
    w = np.zeros(N)
    y = np.zeros(N)
    r = np.array([w_init,0.0],float)
    tol=1.e-5
    for i in range(N):
            w[i]=r[0]
            y[i]=r[1]
            k1=h*f(r,x[i])
            k2=h*f(r+0.5*k1,x[i]+0.5*h)
            k3=h*f(r+0.5*k2,x[i]+0.5*h)
            k4=h*f(r+k3,x[i]+h)
            r+=(k1+2.*k2+2.*k3+k4)/6. #RK4
            if w[i]<1.0:
                x=x[0:i]
                y=y[0:i]
                w=w[0:i]
                break   
    return x, w, y
```
