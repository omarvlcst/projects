# Algoritmo de Verlet

De acuerdo con lo que vimos en clase (y se sacó del PDF de Leap-frog y Verlet), este es el algoritmo para el Verlet de velocidades:

$$v_{n+1/2} = v_n + \frac{1}{2}hF(x_n)$$

$$x_{n+1} = x_n + hv_{n+1/2}$$

Aquí nos interesaría resolver la dinámica newtoniana del problema gravitacional:

$$\ddot{\vec{r}}=-\frac{GM\vec{r}}{\vert\vec{r}\vert^3} \; ,$$ 

es decir:

$$\ddot{x}=-\frac{GMx}{\vert\vec{r}\vert^3}$$

$$\ddot{y}=-\frac{GMy}{\vert\vec{r}\vert^3}$$

```python
GM = 4*pi**2 #con unidades (UA)^3/((Msol)^2*(year))
d = 1 #UA
n = 1 #numero total de revoluciones
N = 365*n #112 puntos son para los cuales la Tierra da una revolucion completa
h = N/8760 #paso de tiempo (1 año/numero total de horas en 1 año)
def verlet(x0,vx0,y0,vy0):
    x = np.zeros(N)
    vx = np.zeros(N)
    y = np.zeros(N)
    vy = np.zeros(N)
    K = np.zeros(N)
    x[0] = x0
    vx[0] = vx0
    y[0] = y0
    vy[0] = vy0
    for i in range(0,N-1):
#        vx[i + 1] = vx[i] - 0.5* h * GM*(x[i]+x[i+1])/d**3
#        x[i + 1] = x[i] + h *vx[i] - 0.5* h**2 * GM*x[i]/d**3
#        vy[i + 1] = vy[i] - 0.5* h * GM*(y[i]+y[i+1])/d**3
#        y[i + 1] = y[i] +  h *vy[i] - 0.5* h**2 * GM*y[i]/d**3
        vx[i+1] = vx[i] - 0.5 * h *  GM*x[i]/d**3
        x[i+1] = x[i] + h * vx[i+1]
        vy[i+1] = vy[i] - 0.5 * h * GM*y[i]/d**3
        y[i+1] = y[i] + h * vy[i+1]
        K[i] += 0.5*(0.5*(vx[i+1]+vx[i]))**2+0.5*(0.5*(vy[i+1]+vy[i]))**2 #-GM/d 
    return x,y,K

xt,yt,Kt = verlet(1,0,0,sqrt(GM/d))

plt.figure(figsize=(7,7))
plt.xlim(-2.,2.)
plt.ylim(-2.,2.)
plt.plot(xt,yt,'.', label='Órbita terrestre')
plt.plot(0,0,'yo', label='Sol')
plt.legend(loc='lower right')
plt.show()
```

```python
t = np.linspace(0,N,N)
u = []
for x in t:
    U = lambda x: -GM/d 
    u.append(U(x))

plt.figure(figsize=(10,5))
plt.title('Energía medida en un lapso de 4 revoluciones')
plt.plot(t,Kt,'.',label='Energía cinética')
plt.plot(t,u,'.',label='Energía potencial')
plt.plot(t,Kt+u,'.',label='Energía total')
plt.xlabel('Tiempo $t$ [1 año/112]')
plt.ylabel('Energía')
plt.legend()
plt.show()
```
